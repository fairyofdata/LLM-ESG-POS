```python
def call_openai_api(input_text):
    openai.api_key = 'sk-'
    response = openai.ChatCompletion.create(
        model="gpt-4-turbo",
        max_tokens=4096,
        messages=[
            {"role": "system", "content": "다음은 ESG를 기준으로 기업의 기사와 보고서 텍스트 데이터를 평가하는 방법에 대해 교수님과 나눈 대화입니다. 맥락에 따라 STT 과정에서 잘못 변환된 내용이라고 판단되는 부분을 고치십시오. 예를 들어, 전혀 맥락에 어울리지 않는 단어가 나오거나, 문법적으로 어색한 표현, 단어 등이 해당합니다."},
            {"role": "user", "content": input_text}
        ]
    )
    print(response)  # 응답 구조를 확인하기 위해 출력
    # 실제 필요한 데이터에 접근하기 전에 응답의 구조를 확인하세요.
    try:
        result = response['choices'][0]['message']['content']
    except KeyError:
        result = "응답에서 필요한 데이터를 찾을 수 없습니다."
    return result
```
